```python
from qiskit import QuantumCircuit

# Define a quantum circuit with 4 qubits (2 for input, 1 for result, 1 for auxiliary)
qc = QuantumCircuit(4)
```

```python
# Addition using Quantum Ripple-Carry Adder
from qiskit.circuit.library import QFT

qc.h(0)
qc.h(1)
qc.append(QFT(3), [0, 1, 2])  # Apply Quantum Fourier Transform
qc.draw('mpl')
```

```python
# Subtraction using Quantum Circuit
qc.x(1)  # Negate the second operand
qc.append(QFT(3), [0, 1, 2])  # Perform addition
qc.draw('mpl')
```

```python
# XOR & AND using Quantum Gates

# 1) XOR using CNOT
qc.cx(0, 1)
qc.draw('mpl')
```

```python
# XOR & AND using Quantum Gates

# 2) AND using Toffoli

qc.ccx(0, 1, 2)  # AND operation
qc.draw('mpl')
```

```python
# --- Multiplication using Controlled Additions ---
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFT

# Define a quantum circuit with 4 qubits (2 inputs, 1 result, 1 auxiliary)
qc = QuantumCircuit(4)  

# Apply Hadamard gates to prepare superposition
qc.h(0)
qc.h(1)

# Apply Quantum Fourier Transform (QFT)
qc.append(QFT(3), [0, 1, 2])  

# Controlled Multiplication (Using Toffoli Gate)
qc.ccx(0, 1, 2)  # Multiply qubits 0 and 1, store result in qubit 2

# Display the circuit
qc.draw('mpl')
```

```python
# --- Division using Inverse QFT and Controlled Subtraction ---
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFT

# Define a quantum circuit with 4 qubits (2 for input, 1 for result, 1 auxiliary)
qc = QuantumCircuit(4)

# Apply Hadamard gates to prepare superposition
qc.h(0)
qc.h(1)

# Apply Quantum Fourier Transform (QFT)
qc.append(QFT(3), [0, 1, 2])  

# Controlled Division (Using Inverse QFT and Toffoli Gate)
qc.append(QFT(3).inverse(), [0, 1, 2])  # Apply Inverse QFT
qc.ccx(0, 1, 2)  # Simulating division by controlled subtraction

# Display the circuit (Use 'text' if Matplotlib is missing)
qc.draw(output='mpl')
```
